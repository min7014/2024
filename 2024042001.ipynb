# Geogebra 영문버전 설치(Install Geogebra English version)

[Windows Installer: GeoGebra Classic 5 Installer for Windows](https://download.geogebra.org/package/win)

위 링크를 선택하여 실행한다.  프로그램이 실행되면, [다음], [동의함], [설치], [마침] 버튼을 차례로 눌러 설치한다.

Select the link above and run it. When the program runs, click the [Next], [I Agree], [Install], and [Finish] buttons in order to install.

image.png

Geogebra가 한글판으로 되어있어도 영문으로 바꿀 수 있습니다. [선택사항], [언어], [E-I], [English(US)] 를 차례로 선택하여 언어를 변경하시면 됩니다.

Even if Geogebra is in Korean, you can change it to English. You can change the language by selecting [선택사항], [언어], [E-I], [English(US)] in that order.

위와 같은 화면이 나오면 설치가 잘 된 것이다.
