- [목성 위성 측정(Jupiter Satellite Measurements)](https://min7014.github.io/2021/2021042619.xls)

[Excel]의 자료는 강원과학고에서 목성 위성 관측하면서 얻은 위성의 시간과 위치에 관한 자료입니다. 다른 수치 자료를 사용하셔도 됩니다.

The data in [Excel] is about the time and location of satellites obtained while observing Jupiter's satellites at Kangwon Science High School. You may use other numerical data.

image.png

image.png

엑셀자료의 수치만 선택하고 Ctrl + C 를 눌러 복사 해놓습니다.

Select only the numbers in the Excel data and press Ctrl + C to copy them.

image.png

A1 셀을 선택하고, Ctrl + v 를 눌러 엑셀수치 자료를 Geogebra의 [SpreadSheet]에 붙여 넣습니다.

Select cell A1 and press Ctrl + V to paste the Excel numeric data into Geogebra's [Spreadsheet].

image.png

Ctrl 누른 상태에서,  순서대로 B열,  A열를 선택하여 데이터를 선택합니다. [Two Variable Regression Analysys] 도구를 선택하면, [Two Variable Regression Analysys] 창이 나타납니다. [Analyze] 선택합니다.

Select data by holding down Ctrl and selecting column B and column A in that order. When you select the [Two Variable Regression Analysys] tool, the [Two Variable Regression Analysys] window appears. Select [Analyze].

image.png

아래 화살표를 선택하면 데이터를 분석할 모델들이 나타납니다. 목성 위성 데이터는  [Sin] 모델를 선택합니다. 그러면 sin 함수를 활용한 회귀모델 식이 자동으로 생성되며 데이터가 분석됨을 볼 수 있습니다. 아마도 자료를 분석하는 과학 분야에서 유용하게 쓸 수 있는 기능일 것입니다.

If you select the arrow below, models for analyzing data will appear. For Jupiter satellite data, select the [Sin] model. Then, you can see that a regression model equation using the sine function is automatically generated and the data is analyzed. This is probably a useful function in the scientific field of analyzing data.
