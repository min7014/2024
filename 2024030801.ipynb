### 교과서 문제

평면 위에 두 정점 $\mathrm{A}$, $\mathrm{B}$가 있고, $\overline{\mathrm{AB}}=2(\sqrt{3}+1)$ 이다. 이 평면 위를 움직이는 세 점 $\mathrm{P}$, $\mathrm{Q}$, $\mathrm{R}$는 항상 $\overline{\mathrm{AP}}=\overline{\mathrm{PQ}}=2$, $\overline{\mathrm{QR}}=\overline{\mathrm{RB}}=\sqrt{2}$를 만족하면서 움직인다. 이 때, 점 $\mathrm{Q}$가 움직이는 영역을 그림으로 나타내고, 그 넓이를 구하여라.
